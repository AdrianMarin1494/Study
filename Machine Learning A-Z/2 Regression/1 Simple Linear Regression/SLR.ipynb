## Ordinary least squares

We need to take our datapoints and project them vertically onto our linear regression line.

yi = the actual amount of potatoes yielded from the farm when that specific amount of nitrogen fertilizer was used

y^i = what the linear regression predicts

We want to find the best line and it will be related to how small the differences are.

Residual is the difference between yi and y^i.
