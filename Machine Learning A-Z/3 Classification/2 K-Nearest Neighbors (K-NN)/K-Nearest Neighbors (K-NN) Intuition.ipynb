Imagine we have a scenario where we have 2 catergories presented in our dataset, one on the left, one on the right.
We add a new data point. Will it be added on the first or second category? For that we use k-nn.

Step 1. Choose the number K of neighbors (most common is 5)

Step 2. Take the K nearest neighbors of the new data point, according to the Euclidean distance (or Manhatten distance, or others)

Step 3. Among these K neighbors, count the number of data points in each category

Step 4. Assign the new data point to the category where you counted the most neighbors
