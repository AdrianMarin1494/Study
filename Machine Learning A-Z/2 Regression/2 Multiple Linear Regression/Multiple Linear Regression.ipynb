## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

We don't have to apply feature scaling. Because in the equation of the multiple linear regression you have the coefficient that is multiplied to each indepenedent variable and it doesen't matter that some features have higher values than others because the coefficient will compensate to put everything on the same scale.

In multiple linear regression there is absolutly no need to apply feature scalling.

We don't need to check the assumptions of linear regression. Whenever you have a new data set and you want to experiment with some machinery models to figure out which one leads to the highest accuracy, even if your data set doesen't have linear relationships you can sitll try a multiple linear regression on it. And if you know your data set doesen't have linear relationships your multiple linear relationships model will just perform poorly, it will get an accuracy lower than the accuracy of your other models, so you will just not select the multiple linear regression model, but you don't have to check the multiple linear regression assumptions. 

## Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Multiple linear regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Thanks to the linear regression class you don't have to worry about the dummy variable trap, selecting the best features meaning the features with the highest P value or the most statistically significant.

## Predicting the Test set reseults

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1))
```
